# <a>Motivação</a>

PCA (Análise de Componentes Principais) é uma ferramente muito utilizada em Machine Learning para reduzir a dimensionalidade de bases, combinando features sem perder a capacidade preditiva do modelo.

Uma das principais aplicações de PCA é no marketing, para segmentação de clientes, usuários, fornecedores, etc. Imagine uma empresa que deseja lançar um novo produto e realiza uma pesquisa para entender a necessidade dos clientes. Esses questionários tem dezenas, e até centenas, de perguntas. Isso pode dificultar a análise, entra a PCA.

# <a>Objeto de Estudo</a>

Os dados disponibilizados pelo Let's Data referem ao lançamento de um novo conceito de veículo, algo parecido com uma microvan (maior que SUV e menor que uma minivan) e a empresa fictícia precisa conhecer melhor o público alvo para iniciar as campanhas de marketing.

O principal método para isso é uma pesquisa direcionada a grupos focais, prosseguindo para pesquisas de média e por fim grande escala. Assim teremos um perfil do público alvo do produto.

O objetivo é, com os dados desses questionários, identificar o perfil ideal do possível comprador para direcionar a estratégia de marketing. Para isso utilizaremos o PCA, pois uma análise de aglomerados (clusterização) com 30 features não teremos bons resultados, precisamos do PCA para reduzir a dimensionalidade.

# <a>Redução de Dimensionalidade - PCA</a>

Temos 30 perguntas, o que seria tranquilo se fossemos aplicar um modelo preditivo, porém queremos segmentar nossos clientes para aplicar diferentes técnicas de marketing para cada grupo. Logo precisamos reduzir essa dimensionalidade, sem perda de muita informação, para tentar encontrar o público alvo para a microvan.

```python
from sklearn.decomposition import PCA
import matplotlib.cm as cm
```

```python
# Removendo as colunas demográficas e a mvliking:
colunas_demograficas = ['age', 'income', 'miles', 'numkids', 'female', 'educ', 'recycle', 'mvliking']

# Testando o nº de componentes e calculando a variância explicada de cada um:
for numero_componentes in range(1, 10):
    # Classse PCA passando o nº de componentes e o df que será reduzido
    pca = PCA(n_components = numero_componentes, random_state=42).fit(df_microvan.drop(colunas_demograficas, axis=1))
    print(f'{numero_componentes} componentes com variâncias: soma de {sum(pca.explained_variance_ratio_)}: {pca.explained_variance_ratio_}')
```

A intenção é escolher o menor número possível que ainda tenha um ganho considerável de variância explicada. Vamos então trabalhar com <b>5 componentes</b> principais.

```python
# Escolhendo 5 componentes principais:
pca = PCA(n_components=5, random_state=42).fit(df_microvan.drop(colunas_demograficas, axis=1))

print(f'Com 5 Componentes Principais temos {sum(pca.explained_variance_ratio_)*100} da variância explicada')
```
